# ソースの準備

```python
with open("aki_recture.txt", "r") as f:
  raw_writedown = f.read()
```

```python
documents = [text.strip() for text in raw_writedown.split(".")]
print("ドキュメントサイズ: ", len(documents))
print("ドキュメントの例: ", documents[10])
```

# リトリートバー

```python
from sentence_transformers import SentenceTransformer

emb_model = SentenceTransformer("infly/inf-retriever-v1-1.5b", trust_remote_code=True)
# In case you want to reduce the maximum length:
emb_model.max_seq_length = 500
```

```python
# Retrievalの実行
question = "akiの強みは？"

query_embeddings = emb_model.encode([question], prompt_name="query")
document_embeddings = emb_model.encode(documents)

# 各ドキュメントの類似度スコア
scores = (query_embeddings @ document_embeddings.T) * 100
print(scores.tolist())
```

```python
topk = 5
for i, index in enumerate(scores.argsort()[0][::-1][:topk]):
  print(f"取得したドキュメント{i+1}: (Score: {scores[0][index]})")
  print(documents[index], "\n\n")
```

```python
from transformers import pipeline
import torch

pipe = pipeline(
    "text-generation",
    model="google/gemma-3-4b-it",
    device="cuda",
    torch_dtype=torch.bfloat16,
    do_sample=False
)
```

# モデル単体での回答

```python
def generate_answer(question):
  message = format("あなたは優秀なAIアシスタントです。以下のストリートファイター6に関する質問に答えてください。\n"\
                 "question: {question}\n"\
                 "answer: ")
  input = message.format(question = question)
  input_len = len(input)

  output = pipe(text_inputs=input, max_new_tokens=500)
  response = output[0]["generated_text"][input_len:]
  return response
```

```python
questions = ["akiの強みは？",
             "akiの弱みは？",
             "akiは投げループできる？",
             "akiの画面端の最大コンボを教えて",
             "akiのシミー後のおすすめのコンボを教えて"]
```

```python
no_RAG_answers = []
for q in questions:
  answer = generate_answer(q)
  no_RAG_answers.append(answer)
```

```python
for i in range(len(no_RAG_answers)):
  print("質問",i+1,": ", questions[i])
  print("\n")
  print("回答",i+1,": ", no_RAG_answers[i])
  print("\n")
```

# RAGを使った回答

```python
#document_embeddings = emb_model.encode(documents)

def _get_references(question, window=0):
  query_embeddings = emb_model.encode([question], prompt_name="query")

  scores = (query_embeddings @ document_embeddings.T) * 100
  if window == 0:
    references = "\n".join(["* " + documents[i] for i in scores.argsort()[0][::-1][:topk]])
  else:
    references = "\n".join(["* " + "。".join(documents[max(0, i-window): min(i+window, len(documents))]).strip() for i in scores.argsort()[0][::-1][:topk]])
  return references


def generate_answer_with_RAG(question, window=0):
  message = format("あなたは優秀なAIアシスタントです。以下のストリートファイター6に関する質問に答えてください。\n"\
                   "また、与えられた資料を参考にして、日本語で回答すること"\
                   "references: {references}"\
                   "question: {question}\n"\
                   "answer: ")

  references = _get_references(question, window)
  input = message.format(question = question, references=references)
  input_len = len(input)

  output = pipe(text_inputs=input, max_new_tokens=500)
  response = output[0]["generated_text"][input_len:]
  return response
```

```python
references = _get_references("akiは投げループできる？")
```

```python
print(references)
```

```python
from tqdm import tqdm

RAG_answers = []
for q in tqdm(questions):
  answer = generate_answer_with_RAG(q)
  RAG_answers.append(answer)
```

```python
for i in range(len(RAG_answers)):
  print("質問",i+1,": ", questions[i])
  print("\n")
  print("回答",i+1,": ", RAG_answers[i])
  print("\n")
```

```python
def translation_en_move_names(answer):
    with open("aki_special_moves.txt", "r") as f:
      special_moves_translation_list = f.read()

    message = format("あなたは優秀なAIアシスタントです。以下の技名の翻訳リストを参考にして、文章中の英語の技名を日本語にしてください\n"\
                   "references: {references}"\
                   "src: {src}\n"\
                   "tgt: ")
    input = message.format(src=answer, references=special_moves_translation_list)
    input_len = len(input)
    output = pipe(text_inputs=input, max_new_tokens=500)
    response = output[0]["generated_text"][input_len:]
    return response
```

```python
def generate_answer_with_RAG_v2(question, window=0):
  with open("aki_special_moves.txt", "r") as f:
      special_moves_translation_list = f.read()

  message = format("あなたは優秀なAIアシスタントです。以下のストリートファイター6に関する質問に答えてください。\n"\
                   "また、与えられた資料を参考にして、日本語で回答すること"\
                   "技名については対応表を参考にして日本語に変換してください\n"\
                   "[special move]: {special_moves_translation_list}\n"\
                   "[references]: {references}\n"\
                   "[question]: {question}\n"\
                   "[answer]: ")

  references = _get_references(question, window)
  input = message.format(question = question, references=references,
                         special_moves_translation_list = special_moves_translation_list)
  input_len = len(input)

  output = pipe(text_inputs=input, max_new_tokens=500)
  response = output[0]["generated_text"][input_len:]

  return response
```

```python
from tqdm import tqdm

RAG_answers_v3 = []
for q in tqdm(questions):
  answer = generate_answer_with_RAG_v2(q)
  RAG_answers_v3.append(answer)
```

```python
for i in range(len(RAG_answers_v3)):
  print("質問",i+1,": ", questions[i])
  print("\n")
  print("回答",i+1,": ", RAG_answers_v3[i])
  print("\n")
```
